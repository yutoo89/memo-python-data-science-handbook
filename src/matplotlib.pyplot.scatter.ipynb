```python
import numpy as np
import matplotlib.pyplot as plt

# 乱数を生成
x = np.random.rand(100)
y = np.random.rand(100)

# 散布図を描画
plt.scatter(x, y)
```

```python
# 固定のステップで増加する数値の散布図を作成
x = np.array(range(100))
print("x shape: ", x.shape)
y = np.array(range(0, 300, 3))
print("y shape: ", y.shape)

plt.scatter(x, y)
```

```python
# 乱数を生成
x = np.random.rand(100)
y = np.random.rand(100)

# サイズ、色、不透明度、色を指定
plt.scatter(
    x,
    y,
    s=60,
    c='pink',
    alpha=0.7,
    edgecolors="red"
)
```

```python
# グラフのタイトル、X 軸、Y 軸の名前 (ラベル)、グリッド線を表示
plt.scatter(x, y)
plt.title("This is a title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
```
